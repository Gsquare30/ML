```python
df.drop(columns=['RowNumber', 'Surname', 'CustomerId'], inplace=True)
```

```python
df.head()
```

```python
df=pd.get_dummies(df, columns=['Geography', 'Gender'], drop_first=True, dtype=int)
df.head()

```

```python
X = df.drop(columns=['Exited'])
y = df['Exited'].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from tensorflow.keras.layers import Dense
```

Keras we have two types of model sequential and non-sequential


```python
model=Sequential()
model.add(Dense(11,activation='relu',input_dim=11))
model.add(Dense(11,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

Model Compilation - decide which loss function and which optimizer to use

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(X_train_scaled,y_train,epochs=100,validation_split=0.2)
```

above cell the epoch value is getting reduced.

```python
history.history
```

```python
model.layers[0].get_weights()
```

33 weights of first layer and 3 weight of 3 biases

```python
y_log=model.predict(X_test_scaled)
```

```python
y_pred=np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
